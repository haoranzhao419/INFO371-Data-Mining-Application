## Question 5 [20 points]: Create two new feature columns: HomeWonLast and VisitorWonLast: if the home team won the previous game, HomeWonLast=1 and VisitorWonLast=0; otherwise, HomeWonLast=0 and VisitorWonLast=1

````python
NBA_df['HomeWonLast'] = 0
NBA_df['VisitorWonLast'] = 0
````

````python
NBA_df.tail(10)
````

````python
from collections import defaultdict
won_last = defaultdict(int)
````

````python
for index, row in NBA_df.sort_values('Date').iterrows():
  home_team = row['Home Team']
  visitor_team = row['Visitor Team']
  NBA_df.at[index, 'HomeWonLast'] = won_last[home_team]
  NBA_df.at[index, 'VisitorWonLast'] = won_last[visitor_team]
  won_last[home_team] = 1 - int(row['VisitorWin'])
  won_last[visitor_team] = int(row['VisitorWin'])
````

````python
NBA_df.tail(10)
````


## Question 6 [10 points]: Apply 5-fold cross validation to fit a decision tree classifier on HomeWonLast and VisitorWonLast against the target VisitorWin. Print out the average accuracy score.

````python
NBA_Team_lastWon = NBA_df[['HomeWonLast', 'VisitorWonLast']]
clf_nba = DecisionTreeClassifier()
scores = cross_val_score(clf_nba, NBA_Team_lastWon, y_true, scoring='accuracy', cv = 5)
print('The average accuracy score is: %s' % np.mean(scores))
````
