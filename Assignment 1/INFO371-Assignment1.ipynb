```python
# import essential libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initizlize a KNN classifier
knn = KNeighborsClassifier()

# Extract the training features
X = df.drop(['gender', 'label'], axis = 1)
X.shape

# Extract the target
y = df['gender']
y.shape
```

```python
# fit the model
knn.fit(X,y)
```

```python
# predict on the features
prediction = knn.predict(X)
```

```python
prediction
```

```python
accuracy_score(y, prediction)
```

7. The default number uses in the neareset neighbors is 5


```python
# Using different numbers of nearest neighbors to fit

## Using K = 1
clf_1 = KNeighborsClassifier(n_neighbors=1)
clf_1.fit(X,y)
prediction_1 = clf_1.predict(X)
accuracy_score(y, prediction_1)
```

```python
## Using K = 2
clf_2 = KNeighborsClassifier(n_neighbors=2)
clf_2.fit(X,y)
prediction_2 =clf_2.predict(X)
accuracy_score(y, prediction_2)
```

```python
## Using K = 3
clf_3 = KNeighborsClassifier(n_neighbors=3)
clf_3.fit(X,y)
prediction_3 = clf_3.predict(X)
accuracy_score(y, prediction_3)
```

## 9.My understand
The number of K affects the the model's preformance. When the number of K become smaller and smaller, the model will be overfitted. Because KNN is a kind of nonparametric model, the prediction result depends on distance between numbers of points. If two points have less distance than another two points, their value of response variable will be closer than another two. When the number of K becomes smaller, the result could easily effect by the local values and increase the effect of noise and make the model easier get a memory of what it learns, that's result in overfitting. And the accuracy score become really high when the number of K is small. 

So in real conditions, we should choose a good number which make the model have a relatively high accuracy score and the computational cost is not very high
